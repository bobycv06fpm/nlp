```python
import nltk
from nltk.grammar import Production, Nonterminal
from nltk import CFG
import copy
```

```python
def handle_long_rules(rules):
    new_rules = list()
    for j in range(len(rules)):
        rhs = list(rules[j].rhs())
        lhs = rules[j].lhs()
        if len(rhs) > 1:
            for i in range(len(rhs) - 1):
                if i == len(rhs) - 2:
                    new_prod = Production(lhs, rhs[i:])
                    new_rules.append(new_prod)
                else:
                    new_sym = Nonterminal(str(j) + str(i))
                    new_prod = Production(lhs, [rhs[i], new_sym])
                    lhs = new_sym
                    new_rules.append(new_prod)
        else:
            new_rules.append(rules[j])
    return new_rules
```
